```
import pandas as pd 
import numpy as np
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
path = "/content/data-preview.csv?url=https:%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_confirmed_global.csv"
data = pd.read_csv(path ,error_bad_lines=False) 
```

```
# Preview the first 5 lines of the loaded data 
data.head()
```

```
#drop unnecessary column
data = data.drop(['Province/State', 'Lat', 'Long'], axis=1)
```

```
data.head(20)
```

```
data = data.groupby(['Country/Region']).sum()
```

```
def parser(x):
	return pd.datetime.strptime(x[:-2] + '20' + x[-2:] , '%m/%d/%Y')

data = pd.read_csv('new_csv.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
data.head()
```
